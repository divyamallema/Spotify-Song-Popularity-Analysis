```
#SKA 


import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

data = pd.read_csv('SpotifyFeatures.csv')
```

```
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import StandardScaler


df = pd.read_csv('SpotifyFeatures.csv')
y = df['popularity']
X = df.drop(columns=['popularity', 'track_name', 'track_id'], axis=1)

for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = LabelEncoder().fit_transform(X[col])

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)

# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


```

```
tree_model = DecisionTreeRegressor(max_depth=12)
tree_model.fit(X_train, y_train)
tree_pred = tree_model.predict(X_test)
tree_MAE = mean_absolute_error(y_test, tree_pred)
tree_MAE
```

```

df = pd.read_csv('SpotifyFeatures.csv')

df_ska = df[df['genre'] == 'Ska']

#original features 
y_ska = df_ska['popularity']
X_ska = df_ska.drop(columns=['popularity', 'track_name', 'track_id', 'genre'], axis=1)

for col in X_ska.columns:
    if X_ska[col].dtype == 'object':
        X_ska[col] = LabelEncoder().fit_transform(X_ska[col])

X_train_ska, X_test_ska, y_train_ska, y_test_ska = train_test_split(X_ska, y_ska, test_size=0.3, random_state=21)

scaler = StandardScaler()
X_train_ska = scaler.fit_transform(X_train_ska)
X_test_ska = scaler.transform(X_test_ska)

tree_model_ska = DecisionTreeRegressor(max_depth=12)
tree_model_ska.fit(X_train_ska, y_train_ska)

tree_pred_ska = tree_model_ska.predict(X_test_ska)
tree_MAE_ska = mean_absolute_error(y_test_ska, tree_pred_ska)

print(f'Decision Tree MAE for Ska genre: {tree_MAE_ska}')
```

```
#ska new model 
df = pd.read_csv('SpotifyFeatures.csv')

df_ska = df[df['genre'] == 'Ska']

# new feautres 
y_ska = df_ska['popularity']
X_ska = df_ska.drop(columns=['popularity', 'track_name', 'track_id', 'genre', 'danceability', 'instrumentalness', 'acousticness', 'key', 'duration_ms', 'speechiness', 'time_signature'], axis=1)

# Apply Label Encoding to categorical variables
for col in X_ska.columns:
    if X_ska[col].dtype == 'object':
        X_ska[col] = LabelEncoder().fit_transform(X_ska[col])

# Split the data into training and test sets
X_train_ska, X_test_ska, y_train_ska, y_test_ska = train_test_split(X_ska, y_ska, test_size=0.3, random_state=21)

# Scale the features
scaler = StandardScaler()
X_train_ska = scaler.fit_transform(X_train_ska)
X_test_ska = scaler.transform(X_test_ska)

# Train the Decision Tree model
tree_model_ska = DecisionTreeRegressor(max_depth=12)
tree_model_ska.fit(X_train_ska, y_train_ska)

# Predict and evaluate the model
tree_pred_ska = tree_model_ska.predict(X_test_ska)
tree_MAE_ska = mean_absolute_error(y_test_ska, tree_pred_ska)

print(f'Decision Tree MAE for Ska genre: {tree_MAE_ska}')
```

```
#new model with all data
df = pd.read_csv('SpotifyFeatures.csv')
y = df['popularity']
X = df.drop(columns=['popularity', 'track_name', 'track_id', 'danceability', 'instrumentalness', 'acousticness', 'key', 'duration_ms', 'speechiness', 'time_signature'], axis=1)

for col in X.columns:
    if X[col].dtype == 'object':
        X[col] = LabelEncoder().fit_transform(X[col])

# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=21)

# Scale the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

tree_model = DecisionTreeRegressor(max_depth=12)
tree_model.fit(X_train, y_train)
tree_pred = tree_model.predict(X_test)
tree_MAE = mean_absolute_error(y_test, tree_pred)
tree_MAE
```
